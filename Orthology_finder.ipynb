```python
import pandas as pd
import urllib.request
import scanpy as sc
import numpy as np
```

```python
met = pd.read_csv('/Users/emilyjones/Desktop/HCA/project/ex_mat/exp1_expression_matrix_annotated.csv')
```

```python
#Download mouse-human homolog information 
# http://www.informatics.jax.org/downloads/reports/index.html#marker
print('Beginning file downloads')

multi_species_homologs =  "http://www.informatics.jax.org/downloads/reports/HOM_AllOrganism.rpt"
homologene_mouse_human = "http://www.informatics.jax.org/downloads/reports/HGNC_homologene.rpt"
orthology_human_mouse = "http://www.informatics.jax.org/downloads/reports/HMD_HumanPhenotype.rpt"
human_mouse_homologs_with_sequence = "http://www.informatics.jax.org/downloads/reports/HOM_MouseHumanSequence.rpt"

url = orthology_human_mouse
urllib.request.urlretrieve(url, './orthology_human_mouse')


#import pandas as pd
df = pd.read_csv("./orthology_human_mouse", sep='\t',index_col=None,header=None)
cols = ['Human Marker'	,'Human Entrez Gene ID',	'Mouse Marker Symbol','HomoloGene ID',	'MGI Marker Accession ID','nan']
df.columns = cols
df["Mouse Marker Symbol"] = df["Mouse Marker Symbol"].str.upper()

# Create a mapping dictionary
human = list(df["Mouse Marker Symbol"])
mouse = list(df["Human Marker"])
dic = dict(zip(mouse,human))
met['HGNC_orth'] = met['symbol'].map(dic)

met
```

```python
met.to_csv('/Users/emilyjones/Desktop/HCA/project/ex_mat/exp_1_ortholog.csv')
```
